```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

import torch
import torch.nn as nn
from sklearn.model_selection import train_test_split
from torch.utils.data import TensorDataset, DataLoader

import helper_functions as help
from models import get_current_model
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'

print(device)
```

```python
model = get_current_model()
model.load_state_dict(torch.load("saved_models/cnn_1.pt"))
model.eval()

model = model.to(device)
print('model', model)
```

```python
'''
Use 32x32 dataset for predicvtions
'''
data, labels = help.read_paths(random_seed=2430, size=32)

#SCALE raw pixel intensities & convert to numpy arrays
data = np.array(data, dtype="float") / 255.0
labels = np.array(labels)

(trainX, testX, trainY, testY) = train_test_split(data, labels, test_size=0.20, random_state=42)


BATCH_SIZE = 16

tensor_testX = torch.Tensor(testX).to(device)
print(tensor_testX.size())
tensor_testX = tensor_testX.permute(0,3,1,2)
tensor_testY = torch.Tensor(testY).to(device)
test_dataset = TensorDataset(tensor_testX, tensor_testY)
test_dataloader = DataLoader(test_dataset, batch_size=BATCH_SIZE)

sample_f, sample_labels = next(iter(test_dataloader))

outputs = model(sample_f)
_, predicted = torch.max(outputs, 1)
```

```python
'''
Read in 128x128 images so can actually see what it is
'''

data, labels = help.read_paths(random_seed=2430, size=128)

#SCALE raw pixel intensities & convert to numpy arrays
data = np.array(data, dtype="float") / 255.0
labels = np.array(labels)

(trainX, testX, trainY, testY) = train_test_split(data, labels, test_size=0.20, random_state=42)

BATCH_SIZE = 16

tensor_testX = torch.Tensor(testX).to(device)
print(tensor_testX.size())
tensor_testX = tensor_testX.permute(0,3,1,2)
tensor_testY = torch.Tensor(testY).to(device)
test_dataset = TensorDataset(tensor_testX, tensor_testY)
test_dataloader = DataLoader(test_dataset, batch_size=BATCH_SIZE)

sample_f, sample_labels = next(iter(test_dataloader))
```

```python
def display_image(dataloader, batch_index=0):
    print('display an image')
    train_features, train_labels = next(iter(dataloader))
    #print(f"Feature batch shape: {train_features.size()}")
    #print(f"Labels batch shape: {train_labels.size()}")
    img = train_features[batch_index].squeeze().cpu()
    # need to flip it b/c opencv reads in images as BGR, matplot reads in as RGB
    img = torch.flip(img,dims=(0,))
    label = train_labels[batch_index]
    plot_img = torch.moveaxis(img,(0,1,2),(-1,0,1))
    #print('plot image shape', plot_img.shape)
    plt.imshow(plot_img)
    plt.show()
    print(f"Label: {label}")
```

```python
batch_index=14

pred = predicted[batch_index]
print("old images = 0, new images = 1")
print(f"Prediction = {pred}")
print(f"True label = {sample_labels[batch_index]}")
if pred == sample_labels[batch_index]:
    print("###############")
    print(f"Correct!")
    print("###############")
else:
    print('wrong prediction')
display_image(test_dataloader, batch_index)
```
